```python
import numpy as np
import pandas as pd
import pickle as pk
import matplotlib.pyplot as plt
import glob
```

```python
pd.set_option('display.max_columns', None)
```

```python
FNS = np.sort(glob.glob('out/opus*pkl'))
```

```python
wv = pd.read_csv('wv.txt', names=['wv'])
wv = wv['wv'].values
```

```python
FNS.shape
```

```python
N = FNS.size

df = np.empty([N, 248])*np.nan

tmp = pd.read_pickle(FNS[0])
for fn in FNS[1:N]:
    tmp = pd.concat([tmp, pd.read_pickle(fn)]);
    # tmp = pd.read_pickle(fn);
    
```

```python
wv_cols = [f"a{w:.1f}" for w in wv[:200]  ]
```

```python
tmp.shape
```

```python

# N = 10501
df = pd.DataFrame(columns=wv_cols +['time_spectrum', 'T_internal', 'Flash_count'], index=tmp.index[:N])


for irow,row in enumerate(tmp['Abs'][0:N]):
    # print(irow, row)
    df.iloc[irow][wv_cols] = row[:200]
    df.iloc[irow]['time_spectrum'] = tmp.iloc[irow]['T_spectrum']
    df.iloc[irow]['T_internal'] = tmp.iloc[irow]['T_internal'][0]
    df.iloc[irow]['Flash_count'] = tmp.iloc[irow]['Flash_count'][0]
```
